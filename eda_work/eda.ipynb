Are there any products with consistently negative profit?

```python
negative_profit_df = df[df["Profit"]< 0]
negative_profit_df.head(3)
```

```python
negative_profit_df["Product Name"].value_counts()
```

```python
df["Product Name"].value_counts()
```

```python
loss_order_perct =  (negative_profit_df["Product Name"].value_counts())/(df["Product Name"].value_counts())*100
loss_order_perct = loss_order_perct.dropna().sort_values(ascending=False)
loss_order_perct
```

```python
consistent_loss_products = pd.DataFrame({"Totalorders":df["Product Name"].value_counts(),
                                        "Lossorders": negative_profit_df["Product Name"].value_counts(),
                                        "perctloss": loss_order_perct})
consistent_loss_products.dropna().sort_values(by = "Totalorders",ascending=False)
```

What’s the return rate (negative profits) by product sub-category?

```python
df.columns
```

```python
df_loss = df[df["Profit"] < 0]
df_loss.head(2)
```

```python
grouped_loss_category = df_loss.groupby(["Sub-Category"])["Profit"].sum().reset_index().sort_values(by="Profit",ascending=True)
grouped_loss_category
```

```python
total_loss = grouped_loss_category["Profit"].sum()
total_loss
```

```python
loss_rate = (grouped_loss_category["Profit"]/grouped_loss_category["Profit"].sum()) * 100
loss_rate
```

```python
loss_rate_category = pd.concat([grouped_loss_category,loss_rate],axis=1)
loss_rate_category.rename(columns={"Profit":"Loss Rate"},inplace=True)
loss_rate_category
```
