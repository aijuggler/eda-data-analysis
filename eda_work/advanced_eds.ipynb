```python
#_____________________________CHATGPT SOLUTION________________________________

# Customers who ordered more than once in same city (ignoring Segment)
customer_city_counts = df.groupby(["Customer ID", "City"]).size()
eligible_pairs_simple = customer_city_counts[customer_city_counts > 1].index

df["eligible_flag_simple"] = pd.Series(list(zip(df["Customer ID"], df["City"]))).isin(eligible_pairs_simple)

print(df["eligible_flag_simple"].value_counts())
df["eligible_flag_simple"]
```

For each Region, find the percentage of orders where shipping took more than 5 days and discount was above 20%.

```python
df[:1]
```

```python
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["shipping_days"] = (df["Ship Date"] - df["Order Date"]).dt.days
df["shipping_days"]
# data filtering where shipping took more than 5 days
data_morethn5days_above20discount = df[df["shipping_days"]>5][df["Discount"]>0.20]
data_morethn5days_above20discount[:2]
```

```python
moretahn5and20 = data_morethn5days_above20discount.groupby("Region")["Order ID"].size()
moretahn5and20
```

```python
total_order_sh = df["Region"].value_counts().sort_index(ascending=True)
total_order_sh
```

```python
order_with5daysshipping_more20discount = (moretahn5and20 / total_order_sh) * 100
order_with5daysshipping_more20discount
```
